Classification Project on Purchasing power depend on several factors

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv('/content/Social_Network_Ads[1].csv')
```

```python
data.head()
```

```python

df=data.drop(['User ID'],axis=1)
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].apply({'Male':1,'Female':0}.get)
```

```python
df.head()
```

```python
x=df.drop(['Purchased'],axis=1)
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
pre=knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,pre)
ac=accuracy_score(y_test,pre)
```
